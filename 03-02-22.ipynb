```python
# CHECK THE GIVEN NUMBER IS PALINDROME OR NOT
# 121--121
# mom--mom
num=int(input("enter a number:"))
temp=num
rev=0
while(num>0):
    dig=num%10
    rev=rev*10+dig
    num=num//10
if(temp==rev):
    print("palindrome")
else:
    print("not palindrome")
```

### Break
- Terminates the current loop statements
- Break is a keyword

```python
for i in "python":
    if i=='h':
        break
    print(i)
```

```python
for s in "ssbsc":
    if s=='h':
        break
    print(s)
```

### continue
- It skips only current itteration and continue the next itterarion
- continue is also a keyword

```python
for i in range(10):
    if i==6:
        continue
        print("ssbsc")
    print(i)
```

```python
# print odd numbers
for i in range(1,9):
    if i%2==0:
        continue
        print("odd number")
    print(i)
```

```python
# print odd numbers
for i in range(-2,9):
    if i%2!=0:
        continue
        print("even number")
    print(i)
```
